```python
# do not change this cell/section
# run this section to initialize your dataset

import random
random.seed(10)
observations = [(random.randrange(0, 10), random.randrange(1, 10), random.randrange(0, 10), \
                random.randrange(0, 10), random.randrange(0, 10)) for i in range(100000)]
```

```python
## Problem  
Imagine that you are a data analyst and your boss wanted you to analyze the dataset `observations` from his survey.
The survey was intended to discover the usual breakfast that most Filipinos wanted on their plate for five consecutive days.
    The dataset was encoded using a list of tuples.  Each tuple stores the order of breakfast of the respondents. 
Coding was used to represent the food options as presented in the table below:

| Code |  Food    |
|------|----------|
|  0   | TapSilog |
|  1   | Tosilog |
|  2   | CornSilog |
|  3   | HotSilog |
|  4   | ChikSilog |
|  5   | BangSilog |
|  6   | Longsilog |
|  7   | Champorado |
|  8   | Arroz Caldo |
|  9   | Tortang Talong |
 
Hence `observations[1]` with the value `(0, 4, 7, 7, 4)` means TapSilog, ChikSilog, Champorado, Champorado, and ChikSilog mix for a 5-day period. 
```

```python
Your goal here is to find the frequent food mix that Filipinos might eat during breakfast.
This will be used by your boss to put up a food business for the public. 
In this exercise, you need to find the frequent food mix.  In your metric, 
you need at least 4 similar responses for the mix to be considered frequent.
```

```python
# Start your code here...
# list of foods used to index the numbers in the list of tuples from observation
# since lists has a index, it will used for mapping the food and the observations
food = [
    "TapSilog",
    "Tosilog",
    "CornSilog",
    "HotSilog",
    "ChikSilog",
    "BangSilog",
    "Longsilog",
    "Champorado",
    "Arroz Caldo",
    "Tortang Talong"
]
```

```python
# creates a dictionary to track all the duplicated data sets
table = {}

# this will check every list of tuples in the dataset and store it inside a dictionary along with the frequency
for item in observations:
    table[item] = table.get(item, 0) + 1

# this will filter the items in the table frequency and retain only the >=4 and store it in the dict called duplicates
duplicates = {k: v for k, v in table.items() if v >= 4}
```

```python
# a statement that prints the frequent food mix 
if duplicates:
    print("\nFrequent food mixes (>=4):")
    for index, (k, v) in enumerate(duplicates.items(), start=1): # a loop that enumarate the duplicates and sets the index to 1
        food_mix = ', '.join(food[index] for index in k if index < len(food)) # this will check the food based on the nubers inside the tuples and match the index in the food and join them using comma  
        print(f"{index}. {k} = {food_mix} \t\t(Count: {v})") # print the food mix with index and the frequency
else:
    print("\nNo duplicates found.")
```
